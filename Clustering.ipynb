```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.cluster.hierarchy as sc
from sklearn.cluster import AgglomerativeClustering
```

```python
df = pd.read_csv('Datasets/Advertising.csv')
```

```python
df.head()
```

```python
data = df[['newspaper', 'sales']]
```

```python
plt.figure(figsize=(10, 7))
sc.dendrogram(sc.linkage(data, metric='euclidean', method='ward'))
plt.title("Agnes")
plt.ylabel("Distances")
plt.show()
```

```python
agnes = AgglomerativeClustering(n_clusters=5, linkage='ward', metric='euclidean')
agnes.fit(data)
```

```python
res_agnes = agnes.fit_predict(data)
```

```python
res_agnes.shape
```

```python
plt.figure(figsize=(10, 7))
sc.dendrogram(sc.linkage(data, metric='euclidean', method='complete'))
plt.title("Diana")
plt.ylabel("Distances")
plt.show()
```

```python
diana = AgglomerativeClustering(n_clusters=5, linkage='complete', metric = 'euclidean')
diana.fit(data)
```

```python
res_diana = diana.fit_predict(data)
```

```python
sns.scatterplot(x='newspaper', y='sales', data=data, hue=agnes.labels_)
plt.show()
```

```python
sns.scatterplot(x=data['newspaper'], y=res_agnes, hue=agnes.labels_)
plt.show()
```

```python
sns.scatterplot(x='newspaper', y='sales', data=data, hue=diana.labels_)
plt.show()
```

```python
sns.scatterplot(x=data['newspaper'], y=res_diana, hue=diana.labels_)
plt.show()
```
